```python
import json
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Read the elevation data
with open('joplin_elevation.json', 'r') as f:
    data = json.load(f)

# Extract elevation data and convert to 2D array
elevation_data = data['results']
rows = len(elevation_data)
cols = len(elevation_data[0])
elevation_array = np.zeros((rows, cols))

# Fill the elevation array
for i in range(rows):
    for j in range(cols):
        elevation_array[i, j] = elevation_data[i][j]['elevation']

# Create a custom colormap for elevation
colors = ['#0000ff', '#00ffff', '#00ff00', '#ffff00', '#ff0000']  # Blue to Red
cmap = LinearSegmentedColormap.from_list('elevation', colors)

# Create the plot
plt.figure(figsize=(12, 8))

# Create contour plot
contour = plt.contourf(elevation_array, cmap=cmap, levels=20)
plt.colorbar(contour, label='Elevation (meters)')

# Add title and labels
plt.title('Joplin Area Elevation Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Save the plot
plt.savefig('joplin_elevation_map.png', dpi=300, bbox_inches='tight')
plt.close()

print("Elevation map saved as 'joplin_elevation.png'")
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import geopandas as gpd

# Read the elevation data
with open('joplin_elevation.json', 'r') as f:
    data = json.load(f)

# Extract elevation data and convert to 2D array
elevation_data = data['results']
rows = len(elevation_data)
cols = len(elevation_data[0])
elevation_array = np.zeros((rows, cols))

# Get the bounds from the elevation data
# The first and last points in the data contain the coordinates
first_point = elevation_data[0][0]
last_point = elevation_data[-1][-1]

xmin = first_point['longitude']
xmax = last_point['longitude']
ymin = first_point['latitude']
ymax = last_point['latitude']

# Fill the elevation array
for i in range(rows):
    for j in range(cols):
        elevation_array[i, j] = elevation_data[i][j]['elevation']

# Read the GeoJSON file
gdf = gpd.read_file('joplin.geojson')

# Create a custom colormap for elevation
colors = ['#0000ff', '#00ffff', '#00ff00', '#ffff00', '#ff0000']  # Blue to Red
cmap = LinearSegmentedColormap.from_list('elevation', colors)

# Create the plot
fig, ax = plt.subplots(figsize=(12, 8))

# Create meshgrid for the contour plot
x = np.linspace(xmin, xmax, cols)
y = np.linspace(ymin, ymax, rows)
X, Y = np.meshgrid(x, y)

# Plot the elevation data
contour = ax.contourf(X, Y, elevation_array, cmap=cmap, levels=20)
plt.colorbar(contour, label='Elevation (meters)')

# Plot the GeoJSON data on top
gdf.plot(ax=ax, facecolor='none', edgecolor='black', linewidth=2)

# Add title and labels
plt.title('Joplin Area Elevation Map with City Boundary')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Set the aspect ratio to equal to maintain proper geographic proportions
ax.set_aspect('equal')

# Save the plot
plt.savefig('joplin_elevation_map_boundary.png', dpi=300, bbox_inches='tight')
plt.close()

print("Elevation map with city boundary saved as 'joplin_elevation_map_boundary.png'")
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import geopandas as gpd
import pandas as pd

# Read the elevation data
with open('joplin_elevation.json', 'r') as f:
    data = json.load(f)

# Extract elevation data and convert to 2D array
elevation_data = data['results']
rows = len(elevation_data)
cols = len(elevation_data[0])
elevation_array = np.zeros((rows, cols))

# Get the bounds from the elevation data
# The first and last points in the data contain the coordinates
first_point = elevation_data[0][0]
last_point = elevation_data[-1][-1]

xmin = first_point['longitude']
xmax = last_point['longitude']
ymin = first_point['latitude']
ymax = last_point['latitude']

# Fill the elevation array
for i in range(rows):
    for j in range(cols):
        elevation_array[i, j] = elevation_data[i][j]['elevation']

# Read the GeoJSON file
gdf = gpd.read_file('joplin.geojson')

# Read the POI CSV file
poi_df = pd.read_csv('../poi_csv.csv')

# Create a custom colormap for elevation
colors = ['#0000ff', '#00ffff', '#00ff00', '#ffff00', '#ff0000']  # Blue to Red
cmap = LinearSegmentedColormap.from_list('elevation', colors)

# Create the plot
fig, ax = plt.subplots(figsize=(12, 8))

# Create meshgrid for the contour plot
x = np.linspace(xmin, xmax, cols)
y = np.linspace(ymin, ymax, rows)
X, Y = np.meshgrid(x, y)

# Plot the elevation data with lower opacity
contour = ax.contourf(X, Y, elevation_array, cmap=cmap, levels=20, alpha=0.5)
plt.colorbar(contour, label='Elevation (meters)')

# Plot the GeoJSON data on top
gdf.plot(ax=ax, facecolor='none', edgecolor='black', linewidth=2)

# Plot POI centroids
ax.scatter(poi_df['centroid_x'], poi_df['centroid_y'], 
           c='red', s=50, alpha=0.7, label='POI Centroids')

# Set axis limits
ax.set_xlim(-94.60, -94.42)
ax.set_ylim(37.00, 37.175)

# Add title and labels
plt.title('Joplin Area Elevation Map with City Boundary and POI Centroids')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Add legend
plt.legend()

# Set the aspect ratio to equal to maintain proper geographic proportions
ax.set_aspect('equal')

# Save the plot
plt.savefig('joplin_elevation_map_boundary_poi.png', dpi=300, bbox_inches='tight')
plt.close()

print("Elevation map with city boundary and POI centroids saved as 'joplin_elevation_map_boundary_poi.png'")
```
